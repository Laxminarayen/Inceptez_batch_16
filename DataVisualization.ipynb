```python
flight_data = pd.read_csv('https://raw.githubusercontent.com/Laxminarayen/Inceptez_batch_16/main/flight_delays.csv',index_col = 'Month')
```

```python
flight_data.tail()
```

```python
plt.figure(figsize= (16,6))
plt.title("Average Arrival delay for Inceptez Airlines WN, by month")

#defenition for the bar 
sns.barplot(x = flight_data.index, y = flight_data['WN'])

#Adding labels 
plt.ylabel("Arrival delay in mins")
plt.xlabel("The months as integers")
```

```python
#Try for NK
```

```python
##Heatmap 
plt.figure(figsize= (8,8))
plt.title("Average Arrival delay for Inceptez Airlines WN, by month")

#defenition for the heatmap - the magnitude of each location monthwise as colours
sns.heatmap(data = flight_data,annot= True,vmin = -10,vmax = 35,cmap='YlGnBu')
#Adding labels 
plt.ylabel("Arrival delay in mins")
plt.xlabel("The months as integers")
```

```python
#Correlation matrix
flight_data.corr()
```

```python
##Heatmap 
plt.figure(figsize= (10,10))
plt.title("Average Arrival delay for Inceptez Airlines WN, by month")

#defenition for the heatmap - the magnitude of each location monthwise as colours
corr = flight_data.corr()
sns.heatmap(data = corr,annot= True,vmin = -1, vmax = 1)
#Adding labels 
plt.ylabel("Arrival delay in mins")
plt.xlabel("The months as integers")
```
